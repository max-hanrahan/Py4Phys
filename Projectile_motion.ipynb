## Ball lightning game

Game to figure out how to hit monsters
* Input position of monster
* Input initial speed
* Get angle(s)
* plot position of ball lighning

```python
# import math
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sqrt(x):
    return np.sqrt(x)
```

```python
# graph y versus x as vert and horizontal coords of ball lightning
# xmin, xmax are min and max x's to plot
def draw_graph(x,y,xmin,xmax):
    plt.plot(x, y, markersize = .8)
    plt.xlabel('x-position')
    plt.ylabel('y-positon')
    plt.title('Trajectory')
    plt.xlim((xmin, 1.5*xmax))
    
```

```python
# here we calculate the actual trajectory
def draw_trajectory(v0, sol):
    g = 9.8
        
    deltx = 1
    delty = 4
    x = np.arange(0.0,deltx, 0.01)
    
    theta1 = np.arctan((deltx + sqrt(deltx**2 - 2*(g*deltx**2 / v0**2)*(g*deltx**2/(2.0*v0**2) + delty)))/(g*deltx**2/v0**2))
    theta2 = np.arctan((deltx - sqrt(deltx**2 - 2*(g*deltx**2 / v0**2)*(g*deltx**2/(2.0*v0**2) + delty)))/(g*deltx**2/v0**2))
        
    theta = 0
    if sol == 1:
        theta = theta1
    elif sol == 2:
        theta = theta2
        
        
    y = math.tan(theta)*x - 0.5*g*x**2/(v0**2*math.cos(theta)**2)
    
    # send draw_graph x and y lists
    draw_graph(x,y,min(x),max(x))
```

```python
# this loop is how we draw sol's for both curves
sols_list = [1,2]
vel = 9

for a in sols_list:
    draw_trajectory(vel, a)
    
plt.legend(['solution1', 'solution2'])
```

```python
vel_list = [10,15,20]
for v in vel_list:
    draw_trajectory(v,1)
plt.legend(['init velocity = 10', 'init_velocity = 15', 'init velocity = 20'])
plt.savefig('./trajectory.pdf', bbox_inches = 'tight')
```
